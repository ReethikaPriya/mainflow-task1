**Correlation between columns**

using scipy

```python
import scipy.stats as stats
import pandas as pd
# Get numeric columns
numeric_cols = dfs.select_dtypes(include=np.number).columns

# Calculate correlation for each pair of numeric columns
correlations = {}
for col1 in numeric_cols:
    correlations[col1] = {}
    for col2 in numeric_cols:
        corr, p_value = stats.pearsonr(dfs[col1], dfs[col2])
        correlations[col1][col2] = corr

# Convert to DataFrame
corr_df = pd.DataFrame(correlations)

# Display the correlation matrix
print(corr_df)
```

**Create a bar chart comparing the average scores of male and female students.**

using numpy

```python
#here code for average scores of gender in all final grades (G1,G2,G3) .
import matplotlib.pyplot as plt
import numpy as np

# Calculate average scores for G1, G2, and G3 for male and female students
grades = ['G1', 'G2', 'G3']
male_avg_scores = [dfs[dfs['sex'] == 'M'][grade].mean() for grade in grades]
female_avg_scores = [dfs[dfs['sex'] == 'F'][grade].mean() for grade in grades]

# Create bar plot
fig, ax = plt.subplots()
bar_width = 0.35
opacity = 0.8
index = np.arange(len(grades))

# Create bars for male and female scores
rects1 = ax.bar(index, male_avg_scores, bar_width,
                alpha=opacity, color='b',
                label='Male')
rects2 = ax.bar(index + bar_width, female_avg_scores, bar_width,
                alpha=opacity, color='y',
                label='Female')

# Set labels and title
ax.set_xlabel('Grade')
ax.set_ylabel('Average Score')
ax.set_title('Average Scores by Gender and Grade')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(grades)
ax.legend()

# Display plot
plt.show()
```
